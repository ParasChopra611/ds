# QQ1] Find out which hour interval sees the most no of orders

KEEP IN MIND THAT THE TIME GIVEN IS IN 24 HR FORMAT

```python
#ist split the Time column into 2 - hour,mins

new_col3 = df["Time"].str.split(':' , n=1,expand = True)
new_col3
```

```python
df["hour"] = new_col3[0]
df["min"] = new_col3[1]
df
```

# Now Heres the twist

the objective here is to find the no of UNIQUE orders that have been registered in each hour interval

so here we wont talk about sales or quantity sold, WHAT WE WILL CONSIDER IS 'ORDER ID' since order id is unique to each order

the problem is that unique order ids are repeated, cuz the entry is done for each item separately

eg suppose i brought 1 battery,1 charger under order_id - 110077 
then the entry is done this way

110077 - 1 battery - $5

110077 - 1 charger - $6

clearly the order id 110077 is repeated 

a timestamp/time interval can have multiple order ids 

but an order id can only have a single unique timestamp

so we will find how many UNIQUE order ids are there in this dataset, we'll drop all rows that have repeated order ids

# no of unique costumers = no of unique orders  = no of unique order ids 

```python
df_uniq_order_id = df.drop_duplicates(subset='Order ID', keep='first') 
# this will remove rows whose order id is repeating, except the 1st
# which means it will only have unique order ids

df_uniq_order_id
```

```python
df_uniq_order_id["Order ID"].value_counts() # just chking if rows with repeating order id is removed or not
```

Nice, see that this new 'df_uniq_order_id' all order id are occured only once

now we can analyse no of unique costumers per hour

lets see the no of unique orders made per each hour

```python
no_uniq_order_by_hour = df_uniq_order_id["hour"].value_counts().sort_index(ascending=True)
no_uniq_order_by_hour
```

```python
plt.figure(figsize=(10,5))
sns.barplot(no_uniq_order_by_hour.index, no_uniq_order_by_hour.values, alpha=0.8)
plt.title('No of unique customers per hour')
plt.ylabel('Number of Unique order ids', fontsize=12)
plt.xlabel('Hour [24 hour format]', fontsize=12)
plt.show()
```

# Clearly 12 hrs [12 pm] to 19 hrs [7 pm] is when the most of the unique customers come for shopping

# Q4] Which time period sees the most sales - Morning, Afternoon or evening?? 
			
			
			Classify the values in 'time' column [created in Q1] in these categories
			
            1st we split time col into hour, min cols [ already did it in ln [28] ]
            
            then we classify hours in morning,afternoon etc
            
			Morning is from sunrise to 11:59 AM. Sunrise typically occurs around 6 AM.
			Afternoon is from 12:01 PM to around 5:00 PM.
			Evening is from 5:01 PM to 8 PM, or around sunset.
			Night is from sunset to sunrise, so from 8:01 PM until 5:59 AM.
			
			
				if hour in range[6 , 12]  => Morning
				if hour in range[12 , 17] => afternoon
				if hour in range[17 , 20] => evening
				if hour in range[20 , 24] => night

```python
df["time_period"] = '' # col that stores morning,afternoon etc

for i in df.index:
    if int(df["hour"].loc[i]) >= 6 and  int(df["hour"].loc[i]) < 12: # 6 am to 11:59 am
        df["time_period"].loc[i] = 'morning'
    
    if int(df["hour"].loc[i]) >= 12 and  int(df["hour"].loc[i]) < 17: #12 pm to 4:59 pm
        df["time_period"].loc[i] = 'afternoon'
        
    if int(df["hour"].loc[i]) >= 17 and  int(df["hour"].loc[i]) < 20: # 5 pm to 7:59 pm
        df["time_period"].loc[i] = 'evening' 
        
    if int(df["hour"].loc[i]) >= 20 and  int(df["hour"].loc[i]) < 24: # 8pm to 11:59 pm
        df["time_period"].loc[i] = 'night'    
    
    if int(df["hour"].loc[i]) >= 0 and  int(df["hour"].loc[i]) < 6: # 12 am to 5:59 am
        df["time_period"].loc[i] = 'night'
df    
```

```python
df.time_period.unique()
```

Nice, we've chked our logic as well

now the actual analysis

```python
sales_by_time_period = df.groupby(['time_period'])['Sales'].sum().reset_index(name ='Total_sales')
sales_by_time_period
```

```python
plt.figure(figsize=(10,5))
sns.barplot(sales_by_time_period.time_period, sales_by_time_period.Total_sales, alpha=0.8)
plt.title('Total sales per time period')
plt.ylabel('Total sales', fontsize=12)
plt.xlabel('Time period', fontsize=12)
plt.show()
```

```python
n_items_by_time_period = df.groupby(['time_period'])['Quantity Ordered'].sum().reset_index(name ='total_quantity_sold')
n_items_by_time_period
```

```python
plt.figure(figsize=(10,5))
sns.barplot(n_items_by_time_period.time_period, n_items_by_time_period["total_quantity_sold"], alpha=0.8)
plt.title('Total quantity sold per time period')
plt.ylabel('Total quantity sold', fontsize=12)
plt.xlabel('Time period', fontsize=12)
plt.show()
```

clearly afternoon sees the most sales while evening sees the least of all

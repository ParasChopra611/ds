each cell is a list of dictionaries, each dict stores a genre with its resp ID [even genres have IDs] in above format

so we'll take each dictionary & retrieve genre 'name' of each dict & append into a single list

but the problem is, those lists are casted as single strings/texts , we need to convert those texts to lists of dicts

below function does the same [below for explanation]

```python
def convert(text):
    List = []
    for i in ast.literal_eval(text): # 
        List.append(i['name']) 
    return List 
```

## Crew Column

```python
movies['crew'].loc[0]
```

we need only director names  

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

## dealing with same word with space

in words like these ---> Sam Worthington, based on novel etc

these are suppose to represent a single entity but they're separated with a space, we need to replace with '_'

```python
def replace_(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ","_"))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(replace_)
movies['crew'] = movies['crew'].apply(replace_)
movies['genres'] = movies['genres'].apply(replace_)
movies['keywords'] = movies['keywords'].apply(replace_)
```

you see words like James Cameron, Sam_Worthington are converted into Sam_Worthington,James_Cameron

# Making recommendations using cosine similarity vals

basically we'll find which movies are similar to given movie by finding similarity b/w their all_details vectors

basically we're finding movies that have similar details to the given movie

## My method

steps

for a given book, find its coresponding vector

then find euclidean dist/cosine simlarity of that vector with other vectors & store them in a dataframe

then sort that dataframe on basis of euclidean dist & display top n vals

```python
given_movie = 'The Lego Movie'
```

finding the index no of this book in dataset, then find the coresponding vector

```python
given_movie_index = movies_new[movies_new['title'] == given_movie].index[0]
given_movie_index
```

```python
# vector for this book

given_movie_vector = vector[given_movie_index]
```

```python
vector[0].shape , given_movie_vector.shape
```

finding euclidean dist/cosine similarity from that given book vector to otgher vectors & store them in a dataframe

```python
from sklearn.metrics.pairwise import cosine_similarity

movie_name = []
similarity_from_given_movie = []

for i in movies_new.index:

  similarity = cosine_similarity(given_movie_vector.reshape(1, -1), vector[i].reshape(1, -1))

  movie_name.append(movies_new['title'].loc[i])
  similarity_from_given_movie.append(similarity)



  #print(books_and_ratings['Book-Title'].loc[i] , similarity)
  
```

```python
similarity_df = pd.DataFrame()
similarity_df['movie_titles'] = movie_name
similarity_df['similarity_from_given_movie'] = similarity_from_given_movie
similarity_df
```

now lets find top n=10 books similar to this book

```python
temp = similarity_df.sort_values(by=['similarity_from_given_movie'],ascending=False)
temp = list(temp['movie_titles'])
temp = temp[0:10]

temp
```

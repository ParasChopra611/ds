[Q1] The given table shows the data science job postings, create a new column "Job level" which tells us if a job is entry,intermediate or senior level.
Also Clean Job title column as well.



```python
data = {'Job Title':['Lead data scientist', 'data scientist [lead]', 'medium level data scientist', 'data scientist entry level','data scientist principal position', 'data scientist','lead data engineer', 'medium level data analyst']}
 
df = pd.DataFrame(data) # Create DataFrame
 
df #original dataframe
```

```python
# Creating new column 'Job level' which stores job lvl of a job ie entry,intermediate or senior/lead

df["Job Level"] = "" #empty new column

for i in df.index:
    if 'lead' in df["Job Title"].loc[i].lower() or 'principal' in df["Job Title"].loc[i].lower(): # if there's lead/principal/senior/sr in job title
        df["Job Level"].loc[i] = "Lead"
        
    elif 'medium' in df["Job Title"].loc[i].lower() or 'intermediate' in df["Job Title"].loc[i].lower():
        df["Job Level"].loc[i] = "Intermediate"
        
    elif 'entry' in df["Job Title"].loc[i].lower():
        df["Job Level"].loc[i] = "Entry Level"
        
        
    else:
        df["Job Level"].loc[i] = 'Not Defined'
        
        
        
df
```

```python
#Now Renaming Job Titles cleaner

for i in df.index:
    if 'science' in df["Job Title"].loc[i].lower() or 'scientist' in df["Job Title"].loc[i].lower(): # if there's science or scioentist in job title
        df["Job Title"].loc[i] = 'Data Scientist' #then replace it with data scientist
        
    elif 'engineer' in df["Job Title"].loc[i].lower():
        df["Job Title"].loc[i] = 'Data Engineer'
        
    elif 'analyst' in df["Job Title"].loc[i].lower():
        df["Job Title"].loc[i] = 'Data Analyst'

df        
```

# This is the final cleaned dataset

[Q2] In given dataset
A] identify the type of salary - per anum/per month/ per week/per hour etc
B] convert all salary types into per anum 

```python
df = pd.DataFrame({"Salary": ['300 per hour', '500 hourly', '600 weekly','700 monthly', '800 yearly', '900 per anum', '1000']})

df
```

```python
# Step1 - create a col "salary type" 

df["Salary Type"] = "" #initialise empty

for i in df.index:
    if 'hour' in df["Salary"].loc[i].lower():
        df["Salary Type"].loc[i] = "Hourly"
        
    if 'week' in df["Salary"].loc[i].lower():
        df["Salary Type"].loc[i] = "Weekly"
        
    if 'month' in df["Salary"].loc[i].lower():
        df["Salary Type"].loc[i] = 'Monthly'
        
    if 'year' in df["Salary"].loc[i].lower() or 'anum' in df["Salary"].loc[i].lower():
        df["Salary Type"].loc[i] = "Yearly"
        


df        
```

```python
# step 2 - clean salary column, remove everything except numbers so that we couls convert them in per anum salary

df['Salary'] = df['Salary'].str.replace(r'[^\d.]+', '')  #code to remove all alphabets from a cell in dataframe
    
    
df    #req ans
```

[Q] in given dataset convert all values to sq ft

```python
df = pd.DataFrame({"Area": ['100 sqft', '100 sq meters', '100sq yards','200 sq meters',]})
df
```

here we'll convert sq meters & sq yards to sq ft
we know that 1 sq yard = 9 sq ft & 1 Square meter= 10.76 square feet

```python
#Step 1 - Creating new column 'Area Unit' which stores the unit of area - sq ft/sq meters/ sq yards etc
df["Area Unit"] = ""

for i in df.index: #for each row in Area column
    if 'meter' in df["Area"].loc[i].lower(): # if theres meter in this cell
        df["Area Unit"].loc[i] = 'Sq. Meters'
    elif 'yards' in df["Area"].loc[i].lower():
        df["Area Unit"].loc[i] = 'Sq. Yards'
    elif 'ft' in df["Area"].loc[i].lower(): 
        df["Area Unit"].loc[i] = 'Sq. Feet'        
        
        
df       
```

```python
# step 2 - removing all alphabets in Area column & convert into numeric values
df['Area'] = df['Area'].str.replace(r'[^\d.]+', '') #removing all alphabets from Area col
df
```

```python
df["Area"] = pd.to_numeric(df["Area"]) # convert this col into numeric tyope so that its divisible/multiplied by a number
df["Area"]
```

```python
# Step 3 - 
#if area unit = sq meters then area* 1/10.76 ie area*0.09293
# if area unit = sq yard then area*1/9  ie area*

#for some reason / by 10.76 & 9 were giving unexpected values so we multiplied by 1/10.76 & 1/9 same thing

for i in df.index:
    if df["Area Unit"].loc[i] == 'Sq. Meters': 
        df["Area"].loc[i] =  df["Area"].loc[i]*0.09293
        
    elif df["Area Unit"].loc[i] == 'Sq. Yards':
        df["Area"].loc[i] =  df["Area"].loc[i]*0.11
        
df        
```

```python
#drop the Area Unit since its of no use anymore & rename Area to Area in Sq. Ft
df = df.drop(['Area Unit'], axis = 1)

df.rename(columns = {'Area':'Area in Sq. Ft'}, inplace = True)
df
```

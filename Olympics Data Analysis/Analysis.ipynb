# Q] There were 2 people who won gold medals in summer & winter season, name them.


```python
# 1st we create 2 lists - participants who won gold in summer season & same with winter season
# then we'll find the intersection of these 2 lists

# A] a list of all participants who won gold in summer season
gold_winner_summer_df = df[(df["Medal"] == 'Gold') & (df["Season"] == 'Summer')]
gold_winner_summer = gold_winner_summer_df["Name"].unique().tolist()
gold_winner_summer
```

```python
# B] a list of same with winter season
gold_winner_winter_df = df[(df["Medal"] == 'Gold') & (df["Season"] == 'Winter')]
gold_winner_winter = gold_winner_winter_df["Name"].unique().tolist()
gold_winner_winter
```

```python
# creating a funct to find the intersection of 2 lists

def intersection(lst1, lst2):
    return list(set(lst1) & set(lst2))
```

```python
intersection(gold_winner_summer, gold_winner_winter)
```
